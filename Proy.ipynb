```julia
using PyPlot
include("Taylor.jl")
using ADT
using ODE
```

```julia
function ec1(t, x̄)
    
    dx = x̄[2]
    dy = (α + β * cos(t)) * sin(x̄[1])
    
    return [dx; dy]
end
```

```julia
x̄ = [0.0, 1.0]
t = linspace(0.0, 4 * pi, 1000)
α = -1.0
β = 0.0

(t1, sol) = ode45(ec1, x̄, t);
```

```julia
x = map(y -> y[1], sol)
y = map(y -> y[2], sol);
```

```julia
plot(x,y)
```
